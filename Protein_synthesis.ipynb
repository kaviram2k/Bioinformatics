```python
DNA_seq='ATGTCCTTTTTATCG'
print('Original Gene: ',DNA_seq)
```

```python
def comlementary_strand(DNA_seq):
  complement_map = str.maketrans("ATGC","TACG")
  complement_gene = DNA_seq.translate(complement_map)
  return complement_gene

complement_gene = comlementary_strand(DNA_seq)
print('cDNA:', complement_gene)
```

```python
def dna_to_mrna(DNA_seq):
  mRNA_seq = complement_gene.replace("T","U")
  return mRNA_seq

mRNA_seq = dna_to_mrna(complement_gene)
print('mRNA:', mRNA_seq)
```

```python
def split_into_codons(mRNA_seq):
  codons = [mRNA_seq[i:i+3] for i in
            range(0, len(mRNA_seq), 3)
            if len(mRNA_seq[i:i+3])==3]
  return codons

codons = split_into_codons(mRNA_seq)
print('Codons:', codons)
```

```python

codon_table = {
    'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
    'UAU': 'Y', 'UAC': 'Y', 'UAA': 'Stop', 'UAG': 'Stop',
    'UGU': 'C', 'UGC': 'C', 'UGA': 'Stop', 'UGG': 'W',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'AUG': 'M',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}

def translate_to_amino_acids(codons):
    amino_acids = "".join(codon_table.get(codon,"?")
                          for codon in codons)
    return amino_acids


amino_acid_seq = translate_to_amino_acids(codon)
print('Amino acids:', amino_acid_seq)


```
